# MISSING VALUES

```python
data={"A":[1,2,3,4,6],"B":[10,40,np.nan,80,60],"C":[np.nan,np.nan,100,50,40]}
```

```python
DF=pd.DataFrame(data)
```

```python
DF
```

```python
DF.isnull()
```

```python
from sklearn.impute import SimpleImputer
```

```python
SM=SimpleImputer(missing_values=np.nan,strategy='mean')
```

```python
SM.fit(DF)
```

```python
SM.transform(DF)
```
